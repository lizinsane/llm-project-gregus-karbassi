## Import Libraries

```python
from docling.document_converter import DocumentConverter


from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
from pathlib import Path
```

## PDF Datei angeben

Gib hier den Pfad zu deiner PDF-Datei an:

```python
# Pfad zu PDF-Datei

#extract = "origin"
extract = "geschichtsbuch"

pdf_path_origin = "docs/Origin by Dan Brown.pdf" 
pdf_path_geschichtsbuch = "docs/Illustrierte Schweizer Geschichte für Sekundar- und Mittelschulen.pdf"  # Passe den Pfad an deine Datei an


if extract == "geschichtsbuch":
    pdf_path = pdf_path_geschichtsbuch 
else:
    pdf_path = pdf_path_origin

if extract == "geschichtsbuch": 
    output_path = "docs/extracted_content_geschichtsbuch.md"
else:
    output_path = "docs/extracted_content_dan_brown.md"
```

## PDF Extraktion

```python
# DocumentConverter initialisieren
converter = DocumentConverter()

# PDF konvertieren
result = converter.convert(pdf_path)

# Ergebnis ausgeben
print("Konvertierung erfolgreich!")
print(f"Anzahl der Seiten: {len(result.document.pages)}")
```

## Text extrahieren

```python
# Text als Markdown exportieren
markdown_text = result.document.export_to_markdown()
print("Extrahierter Text (Markdown-Format):")
print("=" * 50)
print(markdown_text)
```

## Datei speichern

```python
# Als Markdown-Datei speichern
output_path = output_path
with open(output_path, "w", encoding="utf-8") as f:
    f.write(markdown_text)

print(f"Text wurde in '{output_path}' gespeichert")
```

## Erweiterte Optionen

Mit erweiterten Pipeline-Optionen für bessere OCR und Tabellenerkennung:

```python
# Pipeline-Optionen konfigurieren
pipeline_options = PdfPipelineOptions()
pipeline_options.do_ocr = True  # OCR aktivieren für gescannte PDFs
pipeline_options.do_table_structure = True  # Tabellenstruktur erkennen

# Converter mit Optionen erstellen
converter_advanced = DocumentConverter(
    format_options={
        InputFormat.PDF: pipeline_options
    }
)

# PDF mit erweiterten Optionen konvertieren
result_advanced = converter_advanced.convert(pdf_path)

print("Erweiterte Konvertierung abgeschlossen!")
print(f"Anzahl erkannter Tabellen: {sum(1 for item in result_advanced.document.body.items if item.label == 'table')}")
```
